# ***Week7: Model Comparison and Selection***

```python
# Import models:-
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC

#Create models:-
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(min_samples_split=5,min_samples_leaf=5,max_features=None,max_depth=10),
    "Random Forest": RandomForestClassifier(n_estimators=100,random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators = 100,random_state = 42)
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

comparison_list = []

print("--- Running Week 7 Model Selection ---")

for name, model in models.items():
    # Fitting the model
    model.fit(X_train, y_train)

    # Getting predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calculating metrics
    comparison_list.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob)
    })
```

```python
comparison_df = pd.DataFrame(comparison_list).sort_values(by='Accuracy', ascending=False)
print("\nFinal Week 7 Results:")
display(comparison_df)
```
